```python
years=df['iyear'].value_counts().index.tolist()
year_values=df['iyear'].value_counts().values.tolist()
```

```python
akhil = {}
```

```python
for i in np.arange(1,20):
    akhil[years[i]]=year_values[i]
```

```python
a=list(sorted(akhil))
israel=[]
for i in a:
    israel.append(akhil[i])
```

```python
plt.figure(figsize=(15,10))
plt.plot(a,israel)
plt.title("NUMBER OF  ATTACKS PER YEAR",fontsize=23,fontweight="bold")
plt.show()
```
